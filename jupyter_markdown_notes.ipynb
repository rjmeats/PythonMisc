## === <a id="lists"></a> Lists ===============

For **bullet-point** lists, use a dash or asterisk as the first character of the line, followed by a space. Indent to produce nested bullet points (including a different bullet point character at each level). (Alternatively, use HTML list elements instead: &lt;ul&gt; and &lt;li&gt;.)

`- bullet point 1`  
`* bullet point 2`  
`* bullet point 3`  
&nbsp;&nbsp;&nbsp;&nbsp;`* indented bullet point 1`  
&nbsp;&nbsp;&nbsp;&nbsp;`- indented bullet point 2`  

- bullet point 1
* bullet point 2
* bullet point 3
   * indented bullet point 1
   - indented bullet point 2

---

For **numbered lists**, use a number followed by a '.' as the first character of the line, followed by a space. Indent to produce nested lists (including a different numbering/lettering scheme at each level). Note that only the first item at each level needs to use this format - subsequent items in the list can just use an asterisk, and will be numbered automatically. (Alternatively, use HTML list elements instead: &lt;ol&gt; and &lt;li&gt;.)

`1. numbered item 1`  
`2. numbered item 2`    
&nbsp;&nbsp;&nbsp;&nbsp;`1. numbered item 1, second level`  
&nbsp;&nbsp;&nbsp;&nbsp;`* auto numbered item, second level`  
&nbsp;&nbsp;&nbsp;&nbsp;`* auto numbered item, second level`  

1. numbered item 1
2. numbered item 2  
    1. numbered item 1, second level
    * auto numbered item, second level
    * auto numbered item, second level

---

Also note that you need two blank lines between consecutive separate lists to stop them being displayed as one larger list.

## === Links ===============

You can link to external URLs using the format **\[link text\]\(url\)** . Or you can use an HTML **&lt;a href="url"&gt;link text&lt;/a&gt;** element:

For example:

* `This is a link to [Python 3 documentation](https://docs.python.org/3/).`
* `This is also a link to <a href="https://docs.python.org/3/">Python 3 documentation</a>.`


* This is a link to the [Python 3 documentation](https://docs.python.org/3/) site.

___

To specify a link which is internal to the current page, then put an HTML **&lt;a id="xxx"&gt;** element anchor at the relevant point in the current page, and then specify the link as **\[link text\]\(#xxx\)** . 

* `the 'Lists' section heading includes this anchor element <a id="lists"></a>`
* `this is a link to that anchor position: [Link to 'Lists' section](#lists).`

## === Displaying Code ===============

Use text enclosed by single back-tick characters **\`** to display a small section of text highlighted as code:

For example:

"\`This is a single line of code.\`" displays as:

`This is a single line of code.`

and:

"The name of the file is \`filename\`." displays as:

The name of the file is `filename`.

You can also use double back-ticks instead of single back-ticks, which allows a single-back tick to be used in the code text:

"\`\`This is a single \`line\` of code.\`\`" displays as:

``This is a single `line` of code.``

___

You can use an HTML **&lt;code&gt;** element wrapped around text to produce the same effect over multiple lines:
    
<code>
# My function
def f(n) : 
    print("n value is", n)
    s = 0
    for i in range(1, n+1) : 
        s = s + i      
    return s
</code>

___

Blocks of code can be surrounded by triple back-quotes:

 **\`\`\`**<br>
 \[Lines of code\]<br>
 **\`\`\`**<br>

which produces output something like this:

```
# My function
def f(n) : 
    print("n value is", n)
    s = 0
    for i in range(1, n+1) : 
        s = s + i      
    return s
```
___

You can use an HTML **&lt;pre&gt;** element wrapped around code to produce the same effect as triple back-ticks.

___

Add a language name after the opening triple back-ticks for the code section to be given language-sensitive colouring.

 **\`\`\`Python**<br>
 \[Lines of code\]<br>

which produces output something like this:

Simple tables can be produced using the **pipe character '|'** as a column separator and **underscores** as a heading separator, with each row on a separate line.

For example:

```
|Col1|Col2|Col3|
|-----|-------|------|
|Row1 Col1|Row1 Col2|Row1 Col3| 
|Row2 Col1|Row2 Col2|Row2 Col3|
|Row3 Col1|Row3 Col2|Row3 Col3|
|Row4 Col1|Row4 Col2|Row4 Col3|
```

produces:

|Col1|Col2|Col3|
|-----|-------|------|
|Row1 Col1|Row1 Col2|Row1 Col3| 
|Row2 Col1|Row2 Col2|Row2 Col3|
|Row3 Col1|Row3 Col2|Row3 Col3|
|Row4 Col1|Row4 Col2|Row4 Col3|

___

## === Special Symbols ===============

You can display all sorts of special symbols by using their HTML entity name or their Unicode code point number.

For example the Euro symbol can be displayed using:

* via the entity name: `&euro;` &euro;
* via the Unicode number as a decimal value: `&#8364;` &#8364;
* via the Unicode number as a hex value: `&#x20AC;` &#x20AC;

See the [HTML Symbols](https://www.w3schools.com/charsets/ref_utf_punctuation.asp) section of this page for lots of possible symbols and their names/values. A few examples:

* a trademark symbol `&trade;` &trade;
* an up-down arrow `&#x2195;` &#x2195;
* an integral sign `&int;` &int;
* a shaded square `&#x25A8;` &#x25A8;
* an umbrella `&#x2602;` &#x2602;
* a green tick `&#x2705;` &#x2705;
* a neutral face Emoji `&#x1F610;` &#x1F610;

___

You can escape a character's normal markdown meaning by preceding it with a backslash:

`*This is italic* but \*this is not italic\*`

*This is italic* but \*this is not italic\*

## === Mathematical Expressions ===============

Markdown allows Mathematical notation to be displayed by wrapping [LaTex](https://en.wikibooks.org/wiki/LaTeX/Mathematics) expressions with single or double dollar signs:
* within single dollar signs \\$ ... \\$ for inline expressions
* within double dollar signs \\$\\$ ... \\$\\$ for expressions to be displayed centred on their own line

`$\sqrt{k}$ shows a square root expressions within a sentence, using single dollar signs around the LaTex.`

$\sqrt{k}$ shows a square root expressions within a sentence, using single dollar signs around the LaTex.


`But using double dollar signs $$\sqrt{k}$$ displays the square root expression on its own line, centred.`

But using double dollar signs $$\sqrt{k}$$ displays the square root expression on its own line, centred.
